**Ejercicio 2:** Das dos listas encontrar todos los elemtnos que tienen en común

**2.1** Usando bucles

```python
lista1 = [0, 1, 2, 3, 4, 5, 6, 4, 5, 6, 4, 4, 6]
lista2 = [4, 5, 6, 6, 6, 7, 8, 9]

lista = []

for i in lista1:
    if i in lista2 and i not in lista:
        lista.append(lista1[i])
lista
```

**2.2** Usando conjuntos

```python
lista = set(lista1).intersection(set(lista2))
lista
```

**Ejercicio 3:** Dadas dos listas, contar cuantas veces aparece en una de ellas os elementos que tienen en común

```python
lista1 = [0, 1, 2, 3, 4, 5, 6, 4, 5, 6, 4, 4, 6]
lista2 = [4, 5, 6, 6, 6, 7, 8, 9]

elementos = {}

for i in lista1:
    if i in lista2:
        if i in elementos:
            elementos[i] += 1
        else:
            elementos[i] = 1
elementos
```

**Ejercicio 4:** Crear una lista con los primeros 10 múltiplos de 5

**4.1** Usando bucles tradicionales

**Ejercicio 6:** Dada una lista de números encontrear el promedio de los que sean pares

**6.1** Usando solamente estructuras de control

```python
lista1 = [0, 1, 2, 3, 4, 5, 6, 4, 5, 6, 4, 0, 4, 6]

lista = []
suma = 0
n = 0

for i in lista1:
    if lista1[i] % 2 == 0:
        suma = suma + lista1[i]
        n = n+1
print("Promedio: {0}".format(suma/n))
```

**6.2** Usando listas por comprension y funciones incorporadas

```python
lista = [i for i in lista1 if lista1[i] % 2 == 0]
print("Promedio: {0}".format(sum(lista)/len(lista)))
```
